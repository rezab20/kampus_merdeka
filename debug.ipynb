```python
def ubahKeA(X):
  # List huruf vokal.
  vokal = ['a', 'i', 'u', 'e', 'o']
  # Inisialisasi output.
  output = ''
  
  for i in vokal:
    i=i
    for char in X:
      if char in vokal:
          output += i
      else:
          output += char
  #lines = output.split('\n')
  print(output)

lirik = 'balonku ada lima \n'
ubahKeA(lirik)
```

```python
def ubah_vokal(teks):
    vokal_map = {
        'a': 'a',
        'i': 'I',
        'u': 'u',
        'e': 'e',
        'o': 'o'
    }
    result = ""
    for kata in teks.split():
      kata_baru = ""
      for char in kata:
        if char.lower() in vokal_map:
            kata_baru += vokal_map[char.lower()]
        else:
            kata_baru += char
      result += kata_baru + " "
    return result.strip()

lirik = "balonku ada lima"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)

lirik = "balonku ada lima"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)

lirik = "bilinki idi limi"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)

lirik = "bulunku udu lumu"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)

lirik = "belenke ede leme"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)

lirik = "bolonko odo lomo"
hasil_ubah = ubah_vokal(lirik)
print(hasil_ubah)
```
